```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

from sklearn.metrics import silhouette_samples, silhouette_score

# load_iris().keys()
data = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
target = load_iris().target
```

```python
def mapping(d, target):
    dist = 1000
    res = None
    for i in range(len(target)):
#         print(type((d.values-target[i]).T @ (d-target[i])))
#         print(dist)
        if (d.values-target[i]).T @ (d-target[i]) < dist:
            res = i
            dist = (d.values-target[i]).T @ (d-target[i])
    return res
```

```python
sils = []
for n in range(2, 10):
    np.random.seed(0)
    init = data.iloc[list(np.random.choice(range(len(data)), n, replace=False)), :].values
    c = 0
    d = -1
    for _ in range(50):
    #     print(c)
        cluster1 = []
        for i in range(len(data)):
            cluster1.append(mapping(data.iloc[i,:], init))
        temp_d = silhouette_score(data, cluster1)
        if d < temp_d:
            temp = pd.concat([data, pd.DataFrame(np.array(cluster1).reshape(-1, 1), columns=['cl'])], axis=1)
            init = temp.groupby('cl').mean().values
            c += 1
            d = temp_d
        else:
            print(silhouette_score(data, cluster1))
            print(init)
            temp = pd.DataFrame()
            temp['sils'] = silhouette_samples(data, cluster1)
            temp['cl'] = cluster1
            sils.append(temp)
            break
```
