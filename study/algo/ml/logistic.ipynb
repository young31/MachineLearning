```python
import numpy as np
from sklearn.datasets import load_breast_cancer
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

d = load_breast_cancer()

data = pd.DataFrame(d.data, columns=d.feature_names)
target = pd.DataFrame(d.target)

train_X, test_X, train_y, test_y = train_test_split(data, target, test_size=0.3, random_state=2019, shuffle=True)
```

```python
d = (train_X - np.mean(train_X))/np.std(train_X)
data = d.copy()
test_X = (test_X - np.mean(test_X)) / np.std(test_X)
t = test_X.copy()
```

```python
d['const'] = np.ones_like(d['mean area'])
t['const'] = np.ones_like(test_X['mean area'])
```

```python
def logit(x, y, p=False, lr=0.01):
#     if not p:
#         p = np.random.normal(0, 1, len(x[0]))
    t = np.array(list(map(lambda x: 1/(1 + np.exp(-x)), x@p))) - y.flatten()
    
    res = t.T @ x
    return p - np.dot(res, lr/len(x)), np.sum(res)/len(x)
```

```python
np.random.seed(0)
p1 = np.random.normal(0, 0.1, len(d.values[0]))
# p = np.zeros_like(d.columns)
for _ in range(200):
    p1, j = logit(d.values, train_y.values, p1, lr=0.01)
    print(_, '######')
    print(j)
    print(p1)
    if j < 0:
        break
```

```python
def logit_l2(x, y, p=False, lr=0.001, l2=3):
#     if not p:
#         p = np.random.normal(0, 1, len(x[0]))
    t = np.array(list(map(lambda x: 1/(1 + np.exp(-x)), x@p))) - y.flatten()
    
    res = (t.T @ x) / len(x)
    return p - np.dot(res+2*l2*p, lr), np.sum(res)/len(x)
```

```python
np.random.seed(0)
p2 = np.random.normal(0, 0.1, len(d.values[0]))
# p = np.zeros_like(d.columns)
for _ in range(100):
    p2, j = logit_l2(d.values, train_y.values, p2, lr=0.01)
    print(_, '######')
    print(j)
    print(p2)
#     if j < 0:
#         break
```

```python
## 직접 퍼셉트론까지 구현
class Percep:
    def __init__(self, x, y, activation='linear'):
        np.random.seed(0)
        self.data = x
        self.target = y
        self.w = np.random.normal(0, 1, len(x.values[0]))
        self.activation = activation
        self.loss = np.Inf
        self.hist = []
        
    def fit(self):           
        for _ in range(5000):
            w, c = self.logit(self.data, self.target, self.w)
            self.hist.append(c)
            if np.abs(c-self.loss) < 1e-4:
                break
            else:
                self.w = w
                self.loss = c
                
    def predict(self, data):
        return np.array(list(map(lambda x: 1 if x > 0 else 0, data.values @ self.w)))
    
    def logit(self, x, y, w=False, lr=0.01):
        t = np.array(list(map(lambda x: 1/(1 + np.exp(-x)), x@w))) - y.flatten()
        res = t.T @ x
#         np.array(list(map(lambda x: 1 if 1/(1+x) > 0 else 0, t.T @ x)))
        return w + np.dot(res, lr/len(x)), np.sum(res)/len(x)

                       
```
