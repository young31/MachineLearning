# 기존 모델 활용

- 기존 분류모델에 사용한 분류기들을 regression용으로 사용할 수 있음
- 해당 집합의 평균값으로 예측함
- 모형을 생성하는 과정은 기존과 동일

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings(action='ignore')

d = load_boston()

std = StandardScaler()
data = pd.DataFrame(std.fit_transform(d.data), columns=d.feature_names)
target = std.fit_transform(d.target.reshape(-1,1))
dd = data.iloc[:,0:3]
```

```python
from xgboost import XGBRegressor

params= {
    'booster': 'gblinear', # 트리가 아니라 이부분이 바뀐다.
    'tree_method': 'gpu_hist',
    'n_estimators': 400,
    'learning_rate': 0.1,
    'max_depth': 6,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'random_state': 0
}
```

```python
reg = XGBRegressor(**params).fit(data, target)
```

```python
reg.predict(data)
```

```python
from sklearn.neighbors import KNeighborsRegressor

reg_knn = KNeighborsRegressor(n_neighbors=10).fit(data, target)
```

```python
reg_knn.predict(data)
```
