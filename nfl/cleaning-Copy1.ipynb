```python
def stdtype(a):
#     if a == np.nan:
#         return 'NAN'
    a = a.lower()
    if ('ou' in a) or ('open' in a):
        return 'outdoor'
    elif ('in' in a) or ('close' in a) or ('dome' in a) or ('bowl' in a):
        return 'indoor'
    else:
        return a

train['StadiumType'] = train['StadiumType'].replace(np.nan, 'NAN').map(stdtype)
```
