```python
import pandas as pd
import numpy as np

import seaborn as sns

import gc
import os
import time
import warnings
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')

target = train['transaction_real_price']
train = train.drop('transaction_real_price', axis=1)
```

```python
ctd = [
    'transaction_id', 'apartment_id', 'jibun', 'addr_kr', 'apt'
]

train = train.drop(ctd, axis=1)
train['indices'] = np.ones(len(train))
test = test.drop(ctd, axis=1)
test['indices'] = np.ones(len(test))*2
```

```python
merge = pd.concat([train, test], ignore_index=True)
```

```python
merge
```

```python
merge['exclusive_use_area'] = np.log(merge['exclusive_use_area'])
```

```python
x = np.unique(merge['transaction_date'])
def trans_date(a):
    if a == x[0]:
        return 2
    elif a == x[1]:
        return 1
    else:
        return 3
    
merge['transaction_date'] = merge['transaction_date'].map(trans_date)
```

```python
merge['year_of_completion'] = 2019 - merge['year_of_completion']
merge['year_of_completion'] = merge['year_of_completion'] * 12
```

```python
def trans_year(a):
    a = str(a)
    y, m = a[0:4], a[4:]
    res = (2019 - int(y))*12 - int(m)
    return res

merge['transaction_year_month'] = merge['transaction_year_month'].map(trans_year)
```

```python
def n_strip(a):
    for i in range(10):
        a = a.replace(str(i), ' ')
    a = a.split(' ')[0]
    return a

def c_rep(a):
    a = a.replace(' ', '')
    for i in range(6):
        a = a.replace(str(i), '')
    return a

merge['dong'] = merge['dong'].map(n_strip)
```

```python
import autokeras as ak
```

```python
tr = train.drop('indices', axis=1)
te = test.copy()
te.reset_index()

m, s = np.mean(target), np.std(target)
t = (target-m)/s
```

```python
ak_input = ak.StructuredDataInput()
ak_out = ak.StructuredDataBlock(feature_engineering=True,)(ak_input)
ak_out = ak.DenseBlock(use_batchnorm=True, dropout_rate=0.1)(ak_out)
ak_out = ak.RegressionHead()(ak_out)

ak_model = ak.AutoModel(ak_input, ak_out, max_trials=20)
```

```python
import keras

es = keras.callbacks.EarlyStopping(patience=50, restore_best_weights=True)
```

```python
ak_model.fit(tr, t, 
              epochs=1000,
              callbacks=[es],
              validation_split=0.15)
```
