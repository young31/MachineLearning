```python
import pandas as pd
import numpy as np

import glob

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
files = glob.glob('./data/weather_06/*.csv')

dfs = {}
for file in files:
    a = pd.read_csv(file)
    cats = file.split("_")[2]
    
    a.columns = ['day', 'hour', 'forecast', 'value']
    b = a[a['hour']==200].reset_index(drop=True)
    b.fillna(0, inplace=True)
    
    if cats in ['6시간강수량', '6시간적설']:
        res = []
        for i in range(0, b.shape[0], 10):
            c = b.loc[i+1:i+5]['value'].values
            res.append(c)
        temp_df = pd.DataFrame(res, columns=[f'{cats}_' + str(i) for i in range(9, 22, 3)])
        
    elif cats in ['일최저기온']:
        res = []
        for i in range(0, b.shape[0], 2):
            c = min( b.loc[i:i+2]['value'].values)
            res.append(c)
        temp_df = pd.DataFrame(res, columns=[cats])
        
    elif cats in ['일최고기온']:
        res = []
        for i in range(0, b.shape[0], 3):
            c = max(b.loc[i:i+3]['value'].values)
            res.append(c)
        temp_df = pd.DataFrame(res, columns=[cats])
        
    else:
        res = []
        for i in range(0, b.shape[0], 20):
            c = b.loc[i+1:i+5]['value'].values
            res.append(c)

            temp_df = pd.DataFrame(res, columns=[f'{cats}_' + str(i) for i in range(9, 22, 3)])
            
    dfs[cats] = temp_df
    
df2 = pd.concat(list(dfs.values()), axis=1)
```

```python
df1 = pd.concat(list(dfs.values()), axis=1)
```
